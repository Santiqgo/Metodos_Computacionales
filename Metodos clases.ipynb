```python
h=0.01

t=[-np.pi/9]
w=[0]
g=9.77

T=[-np.pi/9]
W=[0]

Te=[-np.pi/9]
O=[0]


ti=[0]

def f(x,t):
    return -g*np.sin(x)
def z(y,t):
    return y


def eu(q,a,b,c):
    return  a+h*q(c,b)

def RK2(q,a,b,c):
    k0=h*q(c,a)
    k1=h*q(c+0.5*c,b+0.5*k0)
    return  b+k1

def RK4(q,a,b,c):
    k0=h*q(c,a)
    k1=h*q(c+h*0.5,b+(0.5*k0))
    k2=h*q(c+h*0.5,b+(0.5*k1))
    k3=h*q(c+h,b+k2)
    return b+(1/6)*(k0+2*k1+2*k2+k3)

for i in range(1,10000):
    t.append(eu(z,t[i-1],w[i-1],ti[i-1]))
    w.append(eu(f,w[i-1],t[i-1],ti[i-1]))
  
    T.append(RK2(z,W[i-1],T[i-1],ti[i-1]))
    W.append(RK2(f,T[i-1],W[i-1],ti[i-1]))

    Te.append(RK4(z,O[i-1],Te[i-1],ti[i-1]))
    O.append(RK4(f,Te[i-1],O[i-1],ti[i-1]))
    ti.append(ti[i-1]+h)
    
plt.plot(t,w,"-.", label="euler")
plt.plot(T,W,"--", label="RK2")
plt.plot(Te,O, label="RK4")
plt.legend()
plt.show()
```
